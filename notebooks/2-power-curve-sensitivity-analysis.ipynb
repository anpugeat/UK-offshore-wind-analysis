# Sensitivity analysis of the power curve to variations of the model parameter

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import cm
matplotlib.rcParams['figure.figsize'] = (10, 5)
import GenericWindTurbinePowerCurve as GWTPC
```

Definition of the reference case:

```python
ref_Drotor  = 80
ref_Pnom    = 2000
ref_Vcutin  = 3
ref_Vcutoff = 25
ref_rMin    = 10
ref_rMax    = 30
ref_CpMax   = 0.4615
ref_TI      = 0.1
ref_Shear   = 0.15
ref_Veer    = 0
ref_Model   = 'Dai et al. 2016'
ref_AirDensity=1.225
ref_ConvEff = 0.92
```

Definition of a vector with wind speed values:

```python
Vws     = np.arange(0,30,0.01)
```

Calculation of the wind turbine power curve corresponding to the reference parameters:

```python
Ref_WTPWC=GWTPC.GenericWindTurbinePowerCurve(Vws,Pnom=ref_Pnom,Drotor=ref_Drotor,Vcutin=ref_Vcutin,Vcutoff=ref_Vcutoff,TI=ref_TI,Shear=ref_Shear,Veer=ref_Veer,rMin=ref_rMin,rMax=ref_rMax,CpMax=ref_CpMax,Model=ref_Model,AirDensity=ref_AirDensity,ConvEff=ref_ConvEff,zhub=100)
```

Definition of the variation intervals:

```python
nb_var = 20
var_Drotor     = np.linspace(40,120,nb_var)
var_Pnom       = np.linspace(1500,2500,nb_var)
var_Vcutin     = np.linspace(0,5,nb_var)
var_Vcutoff    = np.linspace(20,30,nb_var)
var_rMin       = np.linspace(0,15,nb_var)
var_rMax       = np.linspace(15,40,nb_var)
var_CpMax      = np.linspace(0.3,0.59,nb_var) 
var_TI         = np.linspace(0,0.2,nb_var)
var_Shear      = np.linspace(0,0.4,nb_var)
var_Veer       = np.linspace(0,0.5,nb_var)
var_CpModels   = ['Slootweg et al. 2003','Heier 2009','Thongam et al. 2009','De Kooning et al. 2010','Ochieng et Manyonge 2014','Dai et al. 2016','constant']
var_AirDensity = np.linspace(0.5,2,nb_var)
var_ConvEff    = np.linspace(0.8,1,nb_var)
```

## Sensitivity of the power curve to the rotor diameter

```python
var_Drotor  = np.linspace(40,120,nb_var)
ref_Drotor  = 80

plt.figure()
norm = matplotlib.colors.Normalize(vmin=min(var_Drotor), vmax=max(var_Drotor))
sm = plt.cm.ScalarMappable(cmap=cm.jet, norm=plt.Normalize(vmin=min(var_Drotor), vmax=max(var_Drotor)))

for Drotor in var_Drotor:
    rgba_color = np.array(cm.jet(norm(Drotor),bytes=True))/255
    WTPWC=GWTPC.GenericWindTurbinePowerCurve(Vws,Pnom=ref_Pnom,zhub=100,Drotor=Drotor,Vcutin=ref_Vcutin,Vcutoff=ref_Vcutoff,TI=ref_TI,Shear=ref_Shear,Veer=ref_Veer,rMin=ref_rMin,rMax=ref_rMax,CpMax=ref_CpMax,Model=ref_Model,AirDensity=ref_AirDensity,ConvEff=ref_ConvEff)
    plt.plot(Vws,WTPWC,color=rgba_color)

plt.plot(Vws,Ref_WTPWC,'k--',linewidth=2,label='reference case (Drotor=80m)')
plt.grid()
plt.legend()
plt.xlabel('wind speed (m/s)')
plt.ylabel('wind power production (kW)')
plt.colorbar(sm, ax=plt.gca(), label='Rotor diameter (m)')
```
